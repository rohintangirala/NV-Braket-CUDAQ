```python
# speedup from gpu
cpu_runtimes_subset = [cpu_runtimes[size-1] for size in system_size[:2]] 
system_size_subset = system_size[:2]
plt.plot(system_size, p6_gpu8, marker='o', linestyle='-', color='tab:blue', markersize=8, label='8 GPU')
plt.plot(system_size, p6_gpu4, marker='o', linestyle='-', color='tab:orange', markersize=8, label='4 GPU') 
plt.plot(system_size, p6_gpu1, marker='o', linestyle='-', color='tab:green', markersize=8, label='1 GPU')
plt.plot(carbon_nums, cpu_runtimes, marker='o', linestyle='-', color='tab:red', markersize=8, label='CPU') 
plt.title("Ramsey Experiment Benchmarking on CPU and GPUs")
plt.xlabel("Simulated Carbon-13 Nuclei")
plt.ylabel("Runtime (s)")
plt.legend()
plt.savefig("benchmarking_all.png", dpi=300)
```

```python
speedup_gpu8 = []
speedup_gpu4 = []
speedup_gpu1 = []
for i in range(len(system_size_subset)):
    speedup_gpu8.append(cpu_runtimes_subset[i]/p6_gpu8[i])
    speedup_gpu4.append(cpu_runtimes_subset[i]/p6_gpu4[i])
    speedup_gpu1.append(cpu_runtimes_subset[i]/p6_gpu1[i])
speedups_all = {
    '1 GPU': speedup_gpu1,
    '4 GPU': speedup_gpu4,
    '8 GPU': speedup_gpu8
}
```

```python

x = np.arange(len(system_size_subset))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for num_gpu, speedups in speedups_all.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, speedups, width, label=num_gpu)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Execution Time Speedup Ratio (GPU/CPU)')
ax.set_title('Ramsey Experiment Speedup on P6 GPU Instances over CPU')
ax.set_xticks(x + width, system_size_subset)
ax.legend(loc='upper left')
plt.xlabel("Simulated Carbon-13 Nuclei")
# ax.set_ylim(0, 250)
plt.savefig("speedup_all.png", dpi=300)
plt.show()
```
